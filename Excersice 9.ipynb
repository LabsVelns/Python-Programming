###1. Define a class to represent a bank account. Include the following
members: 

Data Members:

a.Name of the Depositor

b.Account Number

c.Type of Account

d.Balance amount in the account

Data Methods:

a.To assign initial values

b.To deposit an amount

c.To withdraw an amount

d.To display name and balance.

Output:
1. New customer
2. Deposit
3. Withdrawal
4. Display
5. Exit

```python

class bank_account:
  def __init__(self,name,accountn,actype,balance):
    self.name = name
    self.accountn = accountn
    self.actype = actype
    self.bal = balance

  def initial_value(self):
    val = self.bal
    print("Initial value - ", val)

  def deposit(self,amt):
    self.amt = amt
    self.bal = self.bal + self.amt
    print("Total balance - ", self.bal)
    print("Current deposit - ", self.amt)

  def withdraw(self,wamt):
    self.wamt = wamt
    if(self.wamt <= self.bal):
      self.bal = self.bal - self.wamt
      print('Current balance - ', self.bal)
      print("Amount withdrawn - ", self.wamt)
    else :
      print("Insufficient Balance")

  def display(self):
    print("Name - ", self.name)
    print("Balance - ", self.bal)

while True:
  x = int(input("\n1. New customer \n2. Deposit \n3. Withdrawal \n4. Display \n5. Exit \n "))

  if (x == 1):
    n = input("Enter your name - ")
    ac = input("Enter account number you want - ")
    at = input("Enter account type (current,saving,salary) - ")
    b =  int(input("Enter amount you want to deposit - "))
    z = bank_account(n,ac,at,b)
    print("----------------------------------------------")
  elif (x == 2):
    dep = int(input("Enter the amount you want to deposit - "))
    z.deposit(dep)
    print("----------------------------------------------")
  elif(x == 3):
    withd = int(input("Enter the amount you want to Withdraw - "))
    z.withdraw(withd)
    print("----------------------------------------------")
  elif(x == 4):
    z.display()
    print("----------------------------------------------")
  elif(x == 5):
    print("----------------------------------------------")
    break
  else:
    print("Invalid Input")
    print("----------------------------------------------")


```

###3. Write program to create a base class staff with code and name. Derive classes teacher(subject , publication) , typist (speed) , officer (grade) . Using the typist class as base class,create two classes regular(salary) and casual(daily wages).Implement a menu driven program for the same.

Output:
1. Teacher
2. Officer,
3. Regular Typist
4. Casaul typist
5. Exit

```python

class staff:
  #Base/Parent Class of code
  def __init__(self,name,code):
    self.name = name
    self.code = code

class teacher(staff):
  #First child/branch class
  def __init__(self,name,code,subject,publication):
    super().__init__(name,code)
    self.sub = subject
    self.pub = publication
  
  def display(self):
    print("Name - ", self.name)
    print("Code - ", self.code)
    print("Subject - ", self.sub)
    print("Publications - ", self.pub)

class officer(staff):
  #First child/branch class
  def __init__(self,name,code,grade):
    super().__init__(name,code)
    self.grade = grade

  def display(self):
    print("Name - ", self.name)
    print("Code - ", self.code)
    print("Grade - ", self.grade)

class typist(staff):
  #First child/branch class
  def __init__(self,name,code,speed):
    super().__init__(name,code)
    self.speed = speed

class t_regular(typist):
  #Second child/branch class of Typist class
  def __init__(self,name,code,speed,salary):
    super().__init__(name,code,speed)
    self.salary = salary

  def display(self):
    print("Name - ", self.name)
    print("Code - ", self.code)
    print("Speed - ", self.speed)
    print("Salary - ", self.salary)

class t_casual(typist):
  #Second child/branch class of Typist class
  def __init__(self,name,code,speed,daily_wages):
    super().__init__(name,code,speed)
    self.wages = daily_wages

  def display(self):
    print("Name - ", self.name)
    print("Code - ", self.code)
    print("Speed - ", self.speed)
    print("Salary - ", self.wages)

p1 = teacher('Nimish','DG001','Python','CDAC Noida')
p2 = officer('Ravi','RP043','S')
p3 = t_regular('Soham','SG099','30',35000)
p4 = t_casual('Vinayak','VC032','28',20000)

while True:
  x = int(input("1. Teacher \n2. Officer \n3. Regular Typist \n4. Casaul typist \n5. Exit \n"))

  if (x == 1):
    p1.display()
    print("----------------------------------------------")
  elif(x==2):
    p2.display()
    print("----------------------------------------------")
  elif(x==3):
    p3.display()
    print("----------------------------------------------")
  elif(x==4):
    p4.display()
    print("----------------------------------------------")
  elif (x == 5):
    print("----------------------------------------------")
    break
  else:
    print("Invalid Choice")
    print("----------------------------------------------")
  
```
