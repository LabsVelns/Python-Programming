###Q1: Write a Program to make a simple calculator that can add, subtract, multiply and divide using functions

```python
def add(x,y):
  print(x+y)
def sub(x,y):
  print(x-y)
def mul(x,y):
  print(x*y)
def div(x,y):
  print(x/y)
a = float(input("Enter 1st number = "))
b = float(input("Enter 2nd number = "))
print("1.Addition \n2.Subtraction\n3.Multiplication\n4.Division\n5.Exit")
while True:
  c = int(input())
  if(c == 1):
    add(a,b)
    print()
  elif (c==2):
    sub(a,b)
    print()
  elif (c==3):
    mul(a,b)
    print()
  elif (c==4):
    div(a,b)
    print()
  elif (c == 5):
    break
  else:
    print("Invalid Input")              
    print()
```

###Q5: Write a Python program to display the Fibonacci sequence up to n-th term by using recursive functions

```python
def fib(x):

  if x<0:
    print("Invalid Input")
  elif(x==0):
    return 0  
  elif x == 1 or x==2:
    return 1
  else:

    return fib(x-1)+fib(x-2)
      
a = int(input("Enter nth term = ")) 
y = fib(a)
print(y)  
```

###Q 6: Write a Python program to find the sum of natural numbers up to n using recursive function

```python
def sum(x):
  if x== 0:
    return 0
  else:
    return x+sum(x-1)

a = int(input("Enter the nth natural number - "))
print(sum(a))    
```

###Q. 10. Write a function find_longest_word() that takes a list of words and returns the length of the longest one.

```python
def lon(x):
  y=len(x)
  l2 = [len(x[i]) for i in range(y)]
  l2.sort()
  print(l2[-1])
  
L1 = list(input("Enter a list of words - ").split())
lon(L1)
```

###Q. 11. Write a function filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer than n

```python
def long(s,y):
  l2 = list(filter(lambda x : (len(x) >= y),s))
  print(l2)

l1 = list(input("Enter list of words - ").split())
n = int(input("Enter the len limit - "))
long(l1,n)
```
