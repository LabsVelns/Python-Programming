###Q9: Write a Python program to add key to a dictionary.
Sample Dictionary : {0: 10, 1: 20}
Expected Result : {0: 10, 1: 20, 2: 30}

```python
dic1={0:10,1:20}
dic1.update(dic1.fromkeys([2],30)) #dic1[key] = value
print(dic1)
```

###Q18: WAP to create a function traiangle to print the following asterisk triangle pattern:

```python
for i in range(3):
  for j in range(i+1) :
    print("*",end= " ")
  print()
for i in range(4):
  for j in range(4-i) :
    print("*",end= " ")
  print() 
```

###Q19: Write a python program to print following multiplication table on the screen

```python
for i in range(1,11):
  print("\t    ",i, end= "")
print()
for i in range(1,32):
  print("___",end="" )
print()
for i in range (1,11):
  print(i,"\t|", end="\t")
  for j in range (1,11):
    print(j*i,end="\t")
  print()  
```
