###6. Write a Python program to write a list content to a file.

```python
x = input("Enter the file name -")
with open(x, mode = "w+") as f:
  l1 = list(input("Enter the content for list - ").split(","))
  f.writelines(l1)
  f.seek(0)
  print(f.read())
```

###7. Write a program to compute the number of characters, words and lines in a file.

```python
 
x=input("Enter the file name - ")

def countLines():
  f = open(x, mode = "r")
  y = f.readlines()
  f.seek(0)
  print(len(y))
  f.close()

def countChar():
  f = open(x, mode = "r")
  f.seek(0)
  print(len(f.read()))
  f.close()

def countWords():
  f = open(x, mode = "r")
  y = f.readlines()
  z = [ i.split(" ") for i in y]
  f.seek(0)
  f.close()
  l2 =[]
  for i in z :
    for j in i :
      l2.append(j)
  print(len(l2))
  f.close()
countLines()
countWords()
countChar()
```
