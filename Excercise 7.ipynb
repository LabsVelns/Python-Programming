###3: Print ten dates, each two a week apart, starting from today, in the form YYYY-MM-DD.

```python
from datetime import datetime,timedelta
n = datetime.now()
a=0
for i in range(1,11):
  
  x = n + timedelta(weeks = a)
  print(f"{i} Date - ",x.strftime("%Y-%m-%d"))
  a+=2
```

###4: Calculate number of days between two given dates
##Given:
 2020-02-25
 
 date_1 = datetime(2020, 2, 25)
 
 2020-09-17

date_2 = datetime(2020, 9, 17)

Expected output:
205 days

```python
import datetime
l1 = list(input("Enter 1st date (YYYY-MM-DD) - ").split("-"))
l2 = list(input("Enter 2nd date (YYYY-MM-DD) - ").split("-"))

x = datetime.datetime(int(l1[0]), int(l1[1]), int(l1[2]))
y = datetime.datetime(int(l1[0]), int(l2[1]), int(l2[2]))
z = x-y
print(z.days," Days")
```

###5: Write a Python script to display the

a) Current date and time

b) Current year in full

c) Month of year full name

d) Weekday of the week

e) Day of year

f) Day of the month

g) Day of week in full name

```python
from datetime import datetime,timedelta
today = datetime.now()
print("Current date and time - ",today)
print("Current year in full - ",today.strftime("%Y"))
print("Month of year full name - ",today.strftime("%B"))
print("Weekday of the week - ",today.strftime("%A"))
#print("Day of year - ",today.strftime("%j"))
jan1 = datetime(2023,1,1)
print("Day of year - ",today-jan1)
print("Day of month - ",today.strftime("%d"))
print("Day of week - ",today.strftime("%w"))
```

###6: Python program to convert a string to datetime:
Jul 1 2016 2:43AM into 2016-07-01 02:43:00

```python
from datetime import datetime
a = input("Enter string of date time - ")
date1 = datetime.strptime(a,"%b %d %Y %I:%M%p")
print(date1)
```
